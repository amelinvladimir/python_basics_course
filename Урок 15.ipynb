```python
print("üö¶ –°–≤–µ—Ç–æ—Ñ–æ—Ä –±—É–¥—É—â–µ–≥–æ")
color = input("–í–≤–µ–¥–∏—Ç–µ —Ü–≤–µ—Ç (–∫—Ä–∞—Å–Ω—ã–π/–∂–µ–ª—Ç—ã–π/–∑–µ–ª–µ–Ω—ã–π): ").lower() # –ë–æ–Ω—É—Å: .lower()

match color:
    case "–∫—Ä–∞—Å–Ω—ã–π":
        print("üõë –°–¢–û–ü")
    case "–∂–µ–ª—Ç—ã–π":
        print("‚ö†Ô∏è –ü—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å—Å—è")
    case "–∑–µ–ª–µ–Ω—ã–π":
        print("‚úÖ –ú–æ–∂–Ω–æ –µ—Ö–∞—Ç—å")
    case _: # Wildcard –¥–ª—è –≤—Å–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤
        print("‚ùå –ù–µ–∏—Å–ø—Ä–∞–≤–Ω–æ—Å—Ç—å —Å–∏–≥–Ω–∞–ª–∞")
```

```python
print("ü§ñ –ì–æ–ª–æ—Å–æ–≤–æ–π –ü–æ–º–æ—â–Ω–∏–∫ (–≤–≤–µ–¥–∏—Ç–µ '–≤—ã—Ö–æ–¥' –¥–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è)")

while True:
    command = input("\n–í–∞—à–∞ –∫–æ–º–∞–Ω–¥–∞: ").lower()
    
    match command:
        case "—Å—Ç–∞—Ä—Ç" | "–Ω–∞—á–∞—Ç—å" | "–≤–∫–ª—é—á–∏":
            print("üöÄ –°–∏—Å—Ç–µ–º–∞ –∑–∞–ø—É—â–µ–Ω–∞")
        case "—Å—Ç–æ–ø" | "–∫–æ–Ω–µ—Ü" | "–≤—ã–∫–ª—é—á–∏":
            print("‚èπÔ∏è –°–∏—Å—Ç–µ–º–∞ –æ—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞")
        case "–∫–∞–∫ –¥–µ–ª–∞" | "—Å—Ç–∞—Ç—É—Å":
            print("ü§ñ –Ø –≤ –Ω–æ—Ä–º–µ")
        case "–ø–æ–º–æ—â—å":
            print("üìú –î–æ—Å—Ç—É–ø–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã: —Å—Ç–∞—Ä—Ç, —Å—Ç–æ–ø, –∫–∞–∫ –¥–µ–ª–∞, –≤—ã—Ö–æ–¥")
        case "–≤—ã—Ö–æ–¥":
            print("üëã –î–æ —Å–≤–∏–¥–∞–Ω–∏—è!")
            break # –í—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞ while
        case _:
            print("ü§∑ –Ø –Ω–µ –ø–æ–Ω–∏–º–∞—é —ç—Ç—É –∫–æ–º–∞–Ω–¥—É")
```

```python
print("üöÅ –ù–∞–≤–∏–≥–∞—Ç–æ—Ä –î—Ä–æ–Ω–∞")

try:
    x = int(input("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ X: "))
    y = int(input("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ Y: "))
    point = (x, y)
    
    match point:
        case (0, 0):
            print("üè† –î—Ä–æ–Ω –Ω–∞ –±–∞–∑–µ")
        case (0, y): # –†–∞—Å–ø–∞–∫–æ–≤–∫–∞: y –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ –∫–æ—Ä—Ç–µ–∂–∞
            print(f"‚¨ÜÔ∏è –î—Ä–æ–Ω –Ω–∞ —Å–µ–≤–µ—Ä–Ω–æ–π –æ—Å–∏ (Y={y})")
        case (x, 0): # –†–∞—Å–ø–∞–∫–æ–≤–∫–∞: x –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ –∫–æ—Ä—Ç–µ–∂–∞
            print(f"‚û°Ô∏è –î—Ä–æ–Ω –Ω–∞ –≤–æ—Å—Ç–æ—á–Ω–æ–π –æ—Å–∏ (X={x})")
        case (x, y):
            print(f"üåç –î—Ä–æ–Ω –≤ —Ç–æ—á–∫–µ ({x}, {y})")
            
except ValueError:
    print("‚ùå –û—à–∏–±–∫–∞: –≤–≤–µ–¥–∏—Ç–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞!")
```
