```python
print("🚦 Светофор будущего")
color = input("Введите цвет (красный/желтый/зеленый): ").lower() # Бонус: .lower()

match color:
    case "красный":
        print("🛑 СТОП")
    case "желтый":
        print("⚠️ Приготовиться")
    case "зеленый":
        print("✅ Можно ехать")
    case _: # Wildcard для всех остальных случаев
        print("❌ Неисправность сигнала")
```

```python
print("🤖 Голосовой Помощник (введите 'выход' для завершения)")

while True:
    command = input("\nВаша команда: ").lower()
    
    match command:
        case "старт" | "начать" | "включи":
            print("🚀 Система запущена")
        case "стоп" | "конец" | "выключи":
            print("⏹️ Система остановлена")
        case "как дела" | "статус":
            print("🤖 Я в норме")
        case "помощь":
            print("📜 Доступные команды: старт, стоп, как дела, выход")
        case "выход":
            print("👋 До свидания!")
            break # Выход из цикла while
        case _:
            print("🤷 Я не понимаю эту команду")
```

```python
# ❌ Старый способ (Много шума)
point = [10, 20]

if isinstance(point, list) and len(point) == 2:
    x = point[0]
    y = point[1]
    print(f"X: {x}, Y: {y}")
elif isinstance(point, list) and len(point) == 3:
    x, y, z = point[0], point[1], point[2]
    print(f"3D Point: {x}, {y}, {z}")
else:
    print("Неизвестный формат")
```

```python
# ✅ Новый способ (Чисто и декларативно)
point = [10, 20]

match point:
    case [x, y]:  # <-- Магия здесь: проверка длины + распаковка
        print(f"2D: X={x}, Y={y}")
    case [x, y, z]:
        print(f"3D: X={x}, Y={y}, Z={z}")
    case _:
        print("Неизвестный формат")
```

```python
print("🚁 Навигатор Дрона")

try:
    x_input = int(input("Координата X: "))
    y_input = int(input("Координата Y: "))
    point = (x_input, y_input)
    match point:
        case (0, 0):
            print("🏠 Дрон на базе")
        case (0, y): # Распаковка: y принимает значение из кортежа
            print(f"⬆️ Дрон на северной оси (Y={y})")
        case (x, 0): # Распаковка: x принимает значение из кортежа
            print(f"➡️ Дрон на восточной оси (X={x})")
        case (x, y):
            print(f"🌍 Дрон в точке ({x}, {y})")
            
except ValueError:
    print("❌ Ошибка: введите целые числа!")
```

```python
flag = False
match (100, 200):
   case (100, 300):  
       print('Case 1')
   case (100, 200) if flag:  
       print('Case 2')
   case (100, 200) if not flag:  
       print('Case 2 not')
   case (100, 200) if flag and True:  
       print('Case 2 not')
   case (100, y):  
       print(f'Case 3, y: {y}')
   case _:  
       print('Case 4')
```

```python
print("🏦 Банковский Фильтр")

operation = input("Тип операции (withdraw/deposit): ").lower()
try:
    amount = float(input("Сумма: "))
    
    match operation:
        case "withdraw":
            if amount > 10000: # Guard можно вынести в case, но здесь для наглядности
                print("⚠️ Требуется подтверждение менеджера")
            else:
                print("💸 Выдача средств")
        case "deposit":
            if amount < 0:
                print("❌ Ошибка: сумма не может быть отрицательной")
            else:
                print("💰 Зачисление")
        case _:
            print("❌ Неизвестная операция")
    
except ValueError:
    print("❌ Ошибка ввода суммы")
```

```python
value = [1, '1']

match value:
    case int():
        print("Это число")
    case str():
        print("Это строка")
    case list():
        print("Это список")
```

```python
# map здесь допустим как встроенная функция конвертации
coords = list(map(int, input("Введите координаты через пробел: ").split()))

match coords:
    case [x, y]:
        print("2D пространство")
    case [x, y, z]:
        print("3D пространство")
    case _:
        print("Ошибка размерности")
```
