```python
price = 1234567.89
print("Цена:", price)
print(f"Цена: {price:,.2f} ₽")
```

```python
print(f"{42}")
print(f"{42:d}") # Десятичное (по умолчанию)
print(f"{42:b}") # Двоичное
print(f"{42:o}") # Восьмеричное
print(f"{42:x}") # Шестнадцатеричное (строчные)
print(f"{42:X}") # Шестнадцатеричное (заглавные)
```

```python
n = 42

# Знак + для положительных чисел
print(f"{n:+d}")   # +42
print(f"{n:+}")   # +42

# Разделитель тысяч
print(f"{1234567:,d}")  # 1,234,567
print(f"{1234567:_d}")  # 1_234_567
```

```python
print(f"{3.14159:.2f}")    # Фиксированная точка
print(f"{123456:.2e}")     # Экспоненциальная (строчные)
print(f"{123456:.3E}")     # Экспоненциальная (заглавные)
print(f"{0.000123:.2g}")   # Автоматический выбор (f или e)
print(f"{0.85:%}")         #
```

```python
x = 42.5

print(f"{x:<10.2f}")  # "42.50      " (влево)
print(f"{x:^10.2f}")  # "  42.50    " (по центру)
print(f"{x:>10.2f}")  # "      42.50" (вправо)
print(f"{x:0>10.2f}") # "0000042.50" (нули слева)
```
