```python
def fib(n):    # записать ряд Фибоначчи меньше n
    """Выведите ряд фибоначчи меньше n."""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()

# Теперь вызовите функцию, которую мы только что определили:
fib(2000)
fib(3000)
fib(4000)
```

```python
def factorial(n):
    i = 1
    f = 1
    while i <= n:
        f *= i
        i += 1
    return f

print(factorial(1))
print(factorial(2))
print(factorial(3))
print(factorial(4))
```

```python
def factorial_list(n):
    i = 1
    f = 1
    result = []
    while i <= n:
        f *= i
        i += 1
        result.append(f)
    return result
    
print(factorial_list(1))
print(factorial_list(2))
print(factorial_list(3))
print(factorial_list(4))
```

```python
def ask_ok(prompt, retries=4, reminder='Please try again!'):
    while True:
        reply = input(prompt)
        if reply in {'y', 'ye', 'yes'}:
            return True
        if reply in {'n', 'no', 'nop', 'nope'}:
            return False
        retries = retries - 1
        if retries == 0:
            raise ValueError('invalid user response')
        print(reminder)
```

```python
ask_ok('Do you really want to quit?')
```

```python
ask_ok('OK to overwrite the file?', 2)
```

```python
def cheeseshop(kind, *arguments, **keywords):
    print("-- У вас есть немного", kind, "?")
    print("-- Сожалею, у нас закончился", kind)
    for arg in arguments:
        print(arg)
    print("-" * 40)
    for kw in keywords:
        print(kw, ":", keywords[kw])
```

```python
cheeseshop("ламбер", "Очень жаль",
           "ОЧЕНЬ ОЧЕНЬ жаль",
           shopkeeper="Michael Palin",
           client="John Cleese",
           sketch="Cheese Shop Sketch")
```
