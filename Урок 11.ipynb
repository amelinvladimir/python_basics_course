```python
rgb = ["Red", "Green", "Blue"]
rgba = rgb

print(id(rgb) == id(rgba))  # they reference the same object
print(rgb is rgba)

rgba.append("Alph")
rgb
```

```python
correct_rgba = rgba[:]
correct_rgba[-1] = "Alpha"
print(correct_rgba)

print(rgba)
```

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
print(letters)

# заменим несколько значений
letters[2:5] = ['C', 'D', 'E']
print(letters)

# удалим несколько значений
letters[2:5] = []
print(letters)

# очищает список, заменив все элементы пустым списком
letters[:] = []
print(letters)
```

```python
fruits = ['apple', 'banana', 'cherry']
cars = ['Ford', 'BMW', 'Volvo']

fruits.extend(cars) # fruits = fruits + cars
# тоже самое, что fruits += cars
print(fruits)
```

```python
print(fruits.pop())
print(fruits)
```

```python
print(fruits.pop(0))
print(fruits)
```

```python
fruits.remove("banana")
print(fruits)
```

```python
fruits.reverse()
print(fruits)
```

```python
fruits.sort(reverse = True)
print(fruits)
```
