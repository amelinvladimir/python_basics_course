```python
# type() позволяет понять тип переменной

i = 100
print(type(i))
```

```python
# input всегда возвращает тип str

i = input()
print(type(i))
print(i + 1)
```

```python
# чтобы преобразовать к типу int используем функцию int()

s = input()
i = int(s)
print(type(i))
print(i + 1)
```

```python
# для удобства записи между цифрами можно писать знак подчеркивания

# эти записи эквивалентны
i = 1_000_000_000_000
j = 1000000000000
k = 10_0000_0_0000_00
print(i)
print(j)
print(k)
```

```python
# числа можно писать в разных системах счисления

i = 0b110010101  # двочная
j = 0o17601      # восьмиричная
k = 0xFa091      # шестнадцатиричная

# вывод по умолчанию в десятиричной системе счисления
print(i)         
print(j)
print(k)
```
