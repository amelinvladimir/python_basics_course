```python
import random

# 1. –ó–∞–≥–∞–¥—ã–≤–∞–µ–º —á–∏—Å–ª–æ
secret_number = random.randint(1, 100)
attempts = 0
guess = None

print("üé≤ –Ø –∑–∞–≥–∞–¥–∞–ª —á–∏—Å–ª–æ –æ—Ç 1 –¥–æ 100. –ü–æ–ø—Ä–æ–±—É–π —É–≥–∞–¥–∞—Ç—å!")

# 2. –¶–∏–∫–ª: –ü–æ–∫–∞ –Ω–µ —É–≥–∞–¥–∞–ª–∏
while True:
    try:
        guess = int(input("–¢–≤–æ–π –≤–∞—Ä–∏–∞–Ω—Ç: "))
        attempts += 1
        
        if guess < secret_number:
            print("üîº –ú–∞–ª–æ! –ë–µ—Ä–∏ –≤—ã—à–µ.")
        elif guess > secret_number:
            print("üîΩ –ú–Ω–æ–≥–æ! –ë–µ—Ä–∏ –Ω–∏–∂–µ.")
        else:
            print(f"üéâ –ü–û–ë–ï–î–ê! –ß–∏—Å–ª–æ –±—ã–ª–æ {secret_number}.")
            print(f"–ü–æ–ø—ã—Ç–æ–∫ –ø–æ—Ç—Ä–∞—á–µ–Ω–æ: {attempts}")
            break
            
    except ValueError:
        print("‚ö†Ô∏è –ù—É–∂–Ω–æ –≤–≤–æ–¥–∏—Ç—å —Ç–æ–ª—å–∫–æ —Ü–∏—Ñ—Ä—ã!")
        # continue —Ç—É—Ç –Ω–µ –Ω—É–∂–µ–Ω, —Ü–∏–∫–ª –ø—Ä–æ—Å—Ç–æ –ø–æ–≤—Ç–æ—Ä–∏—Ç—Å—è
```
