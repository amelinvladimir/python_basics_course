```python
# Система скидок в интернет-магазине (уровень: средний)
# 
# Напишите программу, определяющую размер скидки. 
# У пользователя необходимо запросить: возраст покупателя, общую сумму покупки, наличие скидочной карты, праздник ли сегодня
#
# Условия расчета итоговой скидки:
# 
# 1. Возраст ≥ 60 лет
# +10%
# 2. Сумма заказа ≥ 5000 ₽
# +5%
# 3. Есть дисконтная карта
# +7%
# 4. Праздничный день (is_holiday == True)
# +15%
# Но: максимальная скидка не может превышать 30%
# Программа выводит на экран итоговый процент скидки (целое число от 0 до 30).

age = int(input("Введите возраст покупателя: "))
cost = float(input("Введите сумму заказа: "))
is_disc_cart = input("Есть ли у покупателя скидочная карта? (True/False): ")
is_holiday = input("Праздничный ли сегодня день? (True/False): ")

disc = 0

if age >= 60:
    disc += 10

if cost >= 5000:
    disc += 5

if is_disc_cart == 'True':
    disc += 7

if is_holiday == 'True':
    disc += 15

if disc > 30:
    disc = 30

print(f'Итоговая скидка: {disc}')
print(is_holiday)
```
