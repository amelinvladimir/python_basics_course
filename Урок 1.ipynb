```python
import pandas as pd
from dadata import Dadata
token = "..."
secret = "..."

# Путь к файлу
file_path = r"C:\Users\AmelinVD1\Downloads\адреса1.xlsx"

# Чтение Excel-файла
df = pd.read_excel(file_path, header=None)  # header=None — если нет строки заголовков

# Извлечение первой колонки (индекс 0)
first_column = df.iloc[:, 0]

# Проход по всем значениям в цикле
for value in first_column:
    if pd.notna(value):  # пропускаем пустые ячейки (NaN)
        dadata = Dadata(token, secret)
        result = dadata.clean("address", value)
        print(result)
```
